```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
```

```python
#laad data
data = pd.read_csv("C:/Users/Rania Yunaningrum/OneDrive/Dokumen/SMT 6/DATMIN/Data MBA.csv")
data
```

```python
data = data.drop('NO. INVOICE', axis = 1)
data
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
#nilai support
frequent_itemsets = apriori(data,min_support = 0.85, use_colnames = True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,metric = 'confidence',min_threshold = 0.9)
rules
```

```python
#antaseden length (banyaknya itemset)
rules['antecedents_len'] = rules['antecedents'].apply(lambda x : len(x))
```

```python
mba = rules[(rules['antecedents_len'] >= 1) &
          (rules['confidence']>=0.9) &
          rules['support']>=0.85]
mba
```
